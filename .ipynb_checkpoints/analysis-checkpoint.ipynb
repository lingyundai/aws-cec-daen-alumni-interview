```python
# import aws python sdk
import boto3
import json
import docx
from io import BytesIO

# Get the current session credentials
session = boto3.Session()
credentials = session.get_credentials()

# Print credentials
print("Access Key ID:", credentials.access_key)
print("Secret Access Key:", credentials.secret_key)

import sagemaker

# Get the execution role
role = sagemaker.get_execution_role()
print("Current role:", role)
```

```python
# create S3 client
s3 = boto3.client('s3')

# we have get objects permission in this bucket
bucket_name = 'sagemaker-us-east-1-237863594148'

response = s3.list_objects_v2(Bucket=bucket_name)

# confirm access to all folders and file
for obj in response['Contents']:
    print(obj['Key'])
```

Some transcripts are in the format of JSON, others are in DOCX format. We need to create funtions to read both types.

```python
# read josn file
def read_json_file(bucket, file_name):
    try:
        # get the object from s3
        response = s3.get_object(Bucket=bucket, Key=file_name)
        
        # get the JSON file and access transcript[0] directly
        transcript = json.loads(response['Body'].read().decode('utf-8'))['results']['transcripts'][0]
        print(transcript)
        
        return transcript
    
    except Exception as e:
        print(f"Error reading {file_name}: {e}")
        return None

# read_json_file(bucket_name, "CEC-DAEN-2020-MS-1/CEC-DAEN-2020-MS-1-AUDIO.json")


# read docx file
def read_docx_file(bucket, file_name):
    try:
        # get the object from s3
        response = s3.get_object(Bucket=bucket, Key=file_name)
        
        # create a BytesIO object from the binary content
        doc_binary = BytesIO(response['Body'].read())
        
        # Open as a Word document
        doc = docx.Document(doc_binary)
        print(doc)
        
        # Extract text
        full_text = []
        for paragraph in doc.paragraphs:
            if paragraph.text.strip():
                full_text.append(paragraph.text)
        
        full_text = '\n'.join(full_text)
        print(full_text)
        return full_text
    
    except Exception as e:
        print(f"Error reading {file_name}: {e}")
        return None
    
# read_docx_file(bucket_name, "CEC-DAEN-2019-MS-2/DAEN Interview - Sri.docx")
```

Now we can process the files. To be able to identify the interview the text belongs to, we can initialize a lookup table to map to correct key.

```python
text_look_up = {}

# confirm access to all folders and file
for obj in response['Contents']:
    print(obj['Key'])

text_look_up["CEC-DAEN-2019-MS-1"] = read_docx_file(bucket_name, 
                                                    "CEC-DAEN-2019-MS-1/DAEN Interview - Atabak.docx")
text_look_up["CEC-DAEN-2019-MS-2"] = read_docx_file(bucket_name, 
                                                    "CEC-DAEN-2019-MS-2/DAEN Interview - Sri.docx")
text_look_up["CEC-DAEN-2020-MS-1"] = read_docx_file(bucket_name, 
                                                    "CEC-DAEN-2020-MS-1/CEC-DAEN-2020-MS-1-AUDIO.json")
text_look_up["CEC-DAEN-2020-MS-2"] = read_docx_file(bucket_name, 
                                                    "CEC-DAEN-2020-MS-2/CEC-DAEN-2020-MS-2-AUDIO.json")
text_look_up["CEC-DAEN-2020-MS-2"] = read_docx_file(bucket_name, 
                                                    "CEC-DAEN-2020-MS-2/CEC-DAEN-2020-MS-2-AUDIO.json")
```
